```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats import multitest
```

```python
def t_test(x, y):
    t2, p2 = stats.ttest_ind(x, y)
    if p2 > 0.05:
        print('Нет статистически значимой разницы')
    else:
        print('Есть статистически значимая разница')
```

```python
def func(num):
    list_ex = []
    for i in range(num):
        list_ex.append(np.random.normal(1000,100,100+i))
    for i in range(num-1):
        for j in range(i+1, num):
            print(i, j)
            t_test(list_ex[i], list_ex[j])
```

```python
def mann_whitney(x, y):
    stat, p = stats.mannwhitneyu(x, y)
    if p > 0.05:
        print('Нет статистически значимой разницы')
    else:
        print('Есть статистически значимая разница')
```

```python
def shapiro_test(group):
    W_value,p_value = stats.shapiro(group)
    if p_value > 0.01:
        print('Normal','W=',round(W_value,4),'p-value',round(p_value,4))
    else:
        print('Not normal','W=',round(W_value,4),'p-value',round(p_value,4))
```

```python
s = np.random.lognormal(3, 2, 1000)
shapiro_test(s)
```

### У т-теста проскакивают ошибки, например при сравнении 10 попарно выборок, но меньше чем в Манн-Уитни. И чем больше выборок берем, тем больше накапливается ошибка.

```python
def t_test_with_Bonferron(x, y, comparisons):
    t2, p2 = stats.ttest_ind(x, y)
    if p2 > 0.05/comparisons:
        print('Нет статистически значимой разницы')
    else:
        print('Есть статистически значимая разница')
```

```python
def func(num):
    comp = num*(num-1)/2
    list_ex = []
    for i in range(num):
        list_ex.append(np.random.normal(2+i,3+2*i,5000))
    for i in range(num-1):
        for j in range(i+1, num):
            #print(i, j)
            t_test_with_Bonferron(list_ex[i], list_ex[j], comp)
```

```python
func(20) 
```

# чем больше выборка, тем меньше ошибок (логично), на большое различие ср.квадр тоже реагирует плохо, а на мат.ожидание хорошо(не так много ошибок)

```python
def func(num):
    list_ex = []
    list_p = []
    for i in range(num):
        list_ex.append(np.random.normal(2+i,3+2*i,5000))
    for i in range(num-1):
        for j in range(i+1, num):
            #print(i, j)
            t, p = stats.ttest_ind(list_ex[i], list_ex[j])
            list_p.append(p)
    return list_p
```

```python
def Holm_Bonf(data):
    data = sorted(data)
    count_faults = 0
    for i in data:
        if i < 0.05/(len(data) - (data.index(i) + 1) + 1):
            print ('Не можем  принять гипотезу H0','\ncorrected_alpha=',0.05/(len(data) - (data.index(i) + 1) + 1),
                   '\np_value = ',i)
        else:
            print('Не можем отклонить гипотезу H0','\ncorrected_alpha=',0.05/(len(data) - (data.index(i) + 1) + 1),
                 '\np_value = ',i)
            count_faults+=1
    print('\ncount_faults=', count_faults)
```

```python
p = func(12)
Holm_Bonf(p)
```
